```python
# Importing The Libraries
from skimage.io import imread 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#Loading The Image
image = imread('image.jpg', as_gray=True) 
print('First 10 pixels from first row:', end=' ') 
print(image[0][0:10]) 
plt.title('Original Image'),  
plt.imshow(image, cmap='gray')  
plt.show()
```

```python
# Modifying pixel values to un-normalize image pixels 
normalized_image = image * 255 
print('First 10 pixels from first row:', end=' ') 
print(normalized_image[0][0:10]) 
plt.title('Normalized Pixel Values') 
plt.imshow(normalized_image, cmap='gray') 
plt.show() 
```

```python
# Loading the Image & Accessing Properties 
image = imread('image.jpg') 
print('Image shape:', image.shape) 
print('Total number of pixels:', image.size) 
plt.title('Image') 
plt.imshow(image) 
plt.show()
```

```python
# Loading the Image & Extracting ROI 
image = imread('image.jpg') 
roi = image[800:1000, 1100:1400] 
plt.title('ROI: Corner') 
plt.imshow(roi) 
plt.show() 
```

```python
# Loading Image & Splitting Image Channels 
image = cv2.imread('image.jpg') 
b, g, r = cv2.split(image) 
rgb_image = cv2.merge((r, g, b)) 
# Displaying Different Image Channels 
fig = plt.figure(figsize=(12,10)) 
ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222) 
ax1.imshow(image) 
ax1.title.set_text('BGR Channeled Image') 
ax2.imshow(rgb_image) 
ax2.title.set_text('RGB Channeled Image') 
plt.show()
```

```python
#CW4
img = cv2.imread('home.png')
blue, green, red = cv2.split(img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
# Blue channel
plt.subplot(2, 2, 2)
plt.imshow(blue)
plt.title("Blue Component")
plt.axis("off")
# Green channel
plt.subplot(2, 2, 3)
plt.imshow(green)
plt.title("Green Component")
plt.axis("off")
# Red channel
plt.subplot(2, 2, 4)
plt.imshow(red)
plt.title("Red Component")
plt.axis("off")
# Display the subplot
plt.tight_layout()
plt.show()
```

```python
#Write and execute programs for image arithmetic operations. Obtain different brightness levels by changing the mean value.
```

```python
#loading images
img1 = cv2.imread('lenna.jpg')
img2 = cv2.imread('image.jpg')
height, width = img1.shape[:2]
img2 = cv2.resize(img2, (width, height))
```

```python
#a. Addition
added_image = cv2.add(img1, img2)
plt.imshow(added_image)
```

```python
#b. Subtraction
subtracted_image = cv2.subtract(img1, img2)
plt.imshow(subtracted_image)
```

```python
#c. Mean
mean_value = cv2.mean(img1)[:3]
print(f"Mean value of the image (B, G, R): {mean_value}")
```

```python
#d. Brightening
brightness_offset = 50
brighter_image = cv2.add(img2, np.ones(img2.shape, dtype='uint8') * brightness_offset)
plt.imshow(brighter_image)
```
